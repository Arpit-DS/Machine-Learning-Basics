Trying to know datatypes of Columns in dataframe so that we can make necessary changes.

For ex: Sale Date is an object and we need to convert it into datetime for further operations.

```python
my_df.dtypes
```

It is  very tough to perform operations like Time difference on a string rather a Date Time object.
Changing datatypes of  'SALE DATE'  from object to Python datetime64[ns]. 

So, here I have used to_datetime to convert the expected output which is datetime64[ns]

```python
my_df['SALE DATE'] = pd.to_datetime(my_df['SALE DATE'])
my_df['SALE DATE']
```

Creating a new column "year of sale" by extracting year information from "SALE DATE" column and printing using head command.

Here I have used "df.dt" to access the values of the series as datetime such as "year". The year of the datetime.

```python
my_df['year of sale'] = my_df['SALE DATE'].dt.year
```

```python
my_df.head(5)
```

Dropping 'SALE DATE' from the dataframe by using 
1. df.drop method to drop the "SALE DATE". 
2. axis =1 because we are dropping column.
3. inplace= True for permanently deleting the column

```python
#Dropping column using Column Name
my_df.drop(['SALE DATE'], axis = 1, inplace = True)   

# We can also drop the column using Column Index
#my_df.drop(df.columns[[15]], axis = 1, inplace = True) 

my_df
```

Here we are doing One Hot Encoding using get dummies for 'TAX CLASS AT PRESENT'

The given column already has integer encoding but we nedd to convert it into One hot encoding using get_dummies. the possible reason could be:

Using integer encoding would make model bias the model to assume a natural ordering between categories may result in poor performance or unexpected results (predictions halfway between categories). 

```python
my_df_dummies = pd.get_dummies(my_df, columns=['TAX CLASS AT PRESENT'])
my_df_dummies.head(5)
```
